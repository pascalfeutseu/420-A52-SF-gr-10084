```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.utils.fixes import loguniform
from scipy.stats import randint
```

```python
# Distributions des paramètres
distributions = dict(
    criterion=['gini', 'entropy', 'log_loss'],
    ccp_alpha=loguniform(1e-3, 1e3),
    max_depth=randint(2, 128))

# Estimateur
rnd_forest_cf = RandomForestClassifier(random_state=2023)

# Recherche aléatoire avec avec validation croisée
clf_forest_rnd = RandomizedSearchCV(rnd_forest_cf, 
                                    distributions, 
                                    n_iter=10000, 
                                    cv=5, 
                                    scoring="roc_auc", 
                                    verbose=1, n_jobs=-1, 
                                    random_state=2023)
```

```python
clf_forest_rnd.fit(X_train, y_train)
```
